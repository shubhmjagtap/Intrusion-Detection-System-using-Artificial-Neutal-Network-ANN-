```python
# importing all the packages we will use in this project
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.neural_network import MLPClassifier
%matplotlib inline

# loading the dataset and printin first five values
x_train = pd.read_csv('Train_data.csv')
x_train.head()
```

```python
#checking for number of null values
x_train.isnull().sum()
```

```python
y_train = x_train.iloc[:,-1]
y_train
```

```python
x_train.drop(x_train.columns[[-1,]], axis=1, inplace=True)
```

```python
x_train
```

```python
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
x_train['protocol_type'] = number.fit_transform(x_train['protocol_type'].astype('str'))
x_train
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(30,30,30), random_state=1)
```

```python
clf.fit(x_train, y_train)
```

```python
x_test = pd.read_csv('test_data.csv')
x_test.head()
```

```python
x_test
```

```python
y_test = x_test.iloc[:,-1]
y_test
```

```python
x_test['protocol_type'] = number.fit_transform(x_test['protocol_type'].astype('str'))
x_test
```

```python
x_test.drop(x_test.columns[[-1,]], axis=1, inplace=True)
x_test
```

```python
x_test = x_test.loc[:, ~x_test.columns.str.contains('^Unnamed')]
x_test
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
